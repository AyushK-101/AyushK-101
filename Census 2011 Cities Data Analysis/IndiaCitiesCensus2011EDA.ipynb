**Lets Analyse The Number Of Cities From Each State For Which The Data Is Available**

```python
fig = plt.figure(figsize=(20,10))
states = data.groupby('state_name')['name_of_city'].count().sort_values(ascending=True)
states.plot(kind="barh")
plt.grid(b=True, which='both')
plt.xlabel('Number of cities for each state')
plt.show ()
```

Uttar Pradesh with the most cities, followed by West Bengal. Andaman and Nicobar islands and Nagaland with the least

**Lets compare the states by their literacy rates**

```python
lit_by_states  = data.groupby('state_name').agg({'literates_total': np.sum})
pop_by_states  = data.groupby('state_name').agg({'population_total': np.sum})
literate_rate = lit_by_states.literates_total * 100 / pop_by_states.population_total
literate_rate = literate_rate.sort_values(ascending=True)

plt.subplots(figsize=(15, 10))
ax = sns.barplot(x=literate_rate, y=literate_rate.index)
ax.set_title('States according to literacy rate')
ax.set_xlabel('Literacy Rate as % of population ')
ax.set_ylabel('States', size=15)
```

J&K and UP have the highest literacy rates. Even though literacy rate is high for UP, its citites fell in rank compared to other cities

**Lets plot cities based on population**

```python
def plot_map(sizes, colorbarValue):

    plt.figure(figsize=(19,20))
    f, ax = plt.subplots(figsize=(19, 20))
    map = Basemap(width=5000000, height=3500000, resolution='l', projection='aea', llcrnrlon=69,llcrnrlat=6, urcrnrlon=99, urcrnrlat=36, lon_0=78, lat_0=20, ax=ax)
    map.drawmapboundary()
    map.drawcountries()
    map.drawcoastlines()
    x, y = map(np.array(data["longitude"]), np.array(data["lattitude"]))
    cs = map.scatter(x, y, s=sizes, marker="o", c=sizes, cmap=cm.Dark2, alpha=0.5)
    cbar = map.colorbar(cs, location='right',pad="5%")
    cbar.ax.set_yticklabels(colorbarValue)
    plt.show()
```

```python
population_sizes = data["population_total"].apply(lambda x: int(x /6000))
colorbarValue = np.linspace(data["population_total"].min(), data["population_total"].max(),num=10)
colorbarValue = colorbarValue.astype(int)
plot_map(population_sizes, colorbarValue)
```

**Lets plot female population density**

```python
population_sizes = data["population_female"].apply(lambda x: int(x / 6000))
colorbarValue = np.linspace(data["population_female"].min(), data["population_female"].max(), 
                            num=10)
colorbarValue = colorbarValue.astype(int)

plot_map(population_sizes, colorbarValue)
```
